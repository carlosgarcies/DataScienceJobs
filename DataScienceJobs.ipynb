```python
base_path="/content/drive/MyDrive/archive (1)/job_skills.csv"
df=pd.read_csv(base_path)
```

```python
skills=df['job_skills'].str.split(',', expand=True)
```

```python
skills.head()
```

```python
a=0
lista=[]
for x in range(skills.shape[1]):
  lista.append(set(skills[a].dropna()))
  a+=1
lista[0]
```

```python
lista2=[]
for x in lista:
  for y in x:
      lista2.append(y)
lista3=list(set(lista2))
```

```python
len(lista3)
```

```python
diccionario={}
for x in lista3:
  diccionario[x]=0
```

```python
def sumar(x):
  if x in diccionario.keys():
    diccionario[x]+=1
```

```python
for x in range(skills.shape[1]):
  skills[x].dropna().apply(lambda x: sumar(x))
diccionario
```

```python
totales=pd.DataFrame(diccionario, index=[0])
totales
```

```python
sort=totales.T.sort_values(by=0,ascending=False).head(50)
```

```python
sort=sort.reset_index()
sort=sort.rename(columns={"index":"Skill", 0:"Amount"})
```

```python
sort
```

```python
states=list(da[da["search_country"]=="United States"]["job_location"].str.split(',', expand=True)[1].dropna())
len(states)
```

```python
state_codes=[]
for x in states:
  if len(x)<4:
    state_codes.append(x)
```

```python
count_states_rep={}
for x in state_codes:
  count_states_rep[x]=0
for x in state_codes:
  if x not in count_states_rep.keys():
    pass
  else:
    count_states_rep[x]+=1
```

```python
import plotly.express as px

states = list([x[1:] for x in count_states_rep.keys()])
values = list(count_states_rep.values())

color_scale = ['#ffffff', '#ff0000']


fig = px.choropleth(
    locations=states,
    locationmode="USA-states",
    color=values,
    scope="usa",
    color_continuous_scale=color_scale,
    labels={'color': '# of times in joboard'}
)

fig.update_layout(
    title_text='# of jobs listed per US State'
)
fig.show()
```
